### Question 1 (40%)

In this question, you will model traffic counts in Pittsburgh using Gaussian process (GP) regression.  The included dataset, "PittsburghTrafficCounts.csv", represents the average daily traffic counts computed by traffic sensors at over 1,100 locations in Allegheny County, PA.  The data was collected from years 2012-2014 and compiled by Carnegie Mellon University’s Traffic21 Institute; we have the longitude, latitude, and average daily count for each sensor.  

Given this dataset, your goal is to learn a model of traffic count as a function of spatial location.  To do so, fit a Gaussian Process regression model to the observed data.  While you can decide on the precise kernel specification, you should try to achieve a good model fit, as quantified by a log marginal likelihood value greater than (i.e., less negative than) -1400.  Here are some hints for getting a good model fit:

* We recommend that you take the logarithm of the traffic counts, and then subtract the mean of this vector, before fitting the model.

* Since the data is noisy, don't forget to include a noise term (WhiteKernel) in your model.

* When fitting a GP with RBF kernel on multidimensional data, you can learn a separate length scale for each dimension, e.g., length_scale=(length_scale_x, length_scale_y). 

Your Python code should provide the following five outputs:

1) The kernel after parameter optimization and fitting to the observed data. (10%)

2) The log marginal likelihood of the training data. (5%)

3) Show a 2-D plot of the model's predictions over a mesh grid of longitude/latitude (with color corresponding to the model's predictions) and overlay a 2-D scatter plot of sensor locations (with color corresponding to the observed values). (10%)

4) What percentage of sensors have average traffic counts more than two standard deviations higher or lower than the model predicts given their spatial location? (5%)

5) Show a 2-D scatter plot of the sensor locations, with three colors corresponding to observed values a) more than two standard deviations higher than predicted, b) more then two standard deviations lower than predicted, and c) within two standard deviations of the predicted values. (10%)


```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from sklearn.gaussian_process import GaussianProcessRegressor
from sklearn.gaussian_process.kernels import RBF, ConstantKernel,WhiteKernel
import numpy as np
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.gaussian_process import GaussianProcessClassifier
from sklearn.gaussian_process.kernels import RBF

Data1=pd.read_csv("PittsburghTrafficCounts.csv")
Data1.head()
```

```python
# your code here
Data1['log_trafffic_count'] = np.log(Data1['AvgDailyTrafficCount'])
mean_log_traffic = Data1['log_trafffic_count'].mean()
Data1['log_trafffic_count'] = Data1['log_trafffic_count'] - mean_log_traffic
Data1
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.gaussian_process import GaussianProcessRegressor
from sklearn.gaussian_process.kernels import RBF, WhiteKernel

X = Data1[['Longitude', 'Latitude']].values
y = Data1['log_trafffic_count'].values
```

```python
kernel = 1. * RBF(length_scale=1., length_scale_bounds=(1e-2, 1e2)) \
    + WhiteKernel(noise_level=1.)

gp = GaussianProcessRegressor(kernel=kernel, n_restarts_optimizer=10)
gp.fit(X, y)

x_values = np.linspace(Data1['Longitude'].min(), Data1['Longitude'].max(), 100)
y_values = np.linspace(Data1['Latitude'].min(), Data1['Latitude'].max(), 100)
xx, yy = np.meshgrid(x_values, y_values)
Xr = np.c_[xx.ravel(), yy.ravel()]

y_pred, std_dev = gp.predict(Xr, return_std=True)
Z = y_pred.reshape(xx.shape)
```

```python
plt.figure(figsize=(12, 10))
contour = plt.contourf(xx, yy, Z, cmap='viridis', alpha=0.8)
scatter = plt.scatter(Data1['Longitude'], Data1['Latitude'], c=Data1['log_trafffic_count'], cmap='viridis', edgecolor='k')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Gaussian Process Regression on Traffic Counts')
plt.colorbar(contour, label='Log Traffic Count')
plt.show()
```

```python
Z
```

```python
Data1['Predicted'] = gp.predict(Data1[['Longitude', 'Latitude']].values)
Data1['Z_score'] = (Data1['log_trafffic_count'] - Data1['Predicted'].mean())/Data1['Predicted'].std()
outlier = Data1[abs((Data1['log_trafffic_count'] - Data1['Predicted'].mean())/Data1['Predicted'].std()) > 2.0]
percentage_outliers = (len(outlier) / len(Data1)) * 100
percentage_outliers
```

```python
Data1['Predicted'] = gp.predict(Data1[['Longitude', 'Latitude']].values)
z_scores = (Data1['log_trafffic_count'] - Data1['Predicted'].mean())/Data1['Predicted'].std()
colors = np.where(z_scores > 2, 'red', np.where(z_scores < -2, 'blue', 'green'))

# Create the scatter plot
plt.figure(figsize=(12, 10))
plt.scatter(Data1['Longitude'], Data1['Latitude'], c=colors, edgecolor='k', s=50)
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Sensor Locations Classified by Deviation from Predicted Traffic Counts')
plt.grid(True)
plt.show()
```

### Part 1: 
Learn a Bayesian network structure using only the six features ["'\% Smokers'","'\% Obese'","'Violent Crime Rate'","'80/20 Income Ratio'","'\% Children in Poverty'","'Average Daily PM2.5'"].  Use pd.cut() to discretize each feature into 5 categories: 0,1,2,3,4. 

```python
# Discretize each feature into 5 categories. Ensure labels are strings if concatenation is involved later.
# Define the features with correct column names
features = ["County","'\% Smokers'","'\% Obese'","'Violent Crime Rate'","'80/20 Income Ratio'","'\% Children in Poverty'","'Average Daily PM2.5'"]

# Create Data3 from Data2 with the specified columns
Data3 = Data2[features]
for col in Data3.columns[1:]:  
    Data3[col] = pd.to_numeric(Data3[col], errors='coerce')

Data3
```

```python
for col in Data3.iloc[:, 1:]: 
    Data3[col] = pd.cut(Data3[col], bins=5, labels=[0, 1, 2, 3, 4])
Data3
```

#### (a) Use HillClimbSearch and BicScore to learn the Bayesian network structure (5%)

```python
from pgmpy.estimators import HillClimbSearch
from pgmpy.estimators import BicScore
# For other versions of pgmpy, move scoring_method parameter from initialization function to estimate function 
hc = HillClimbSearch(Data3)
best_model = hc.estimate(scoring_method='BicScore')
print(best_model.edges())
```

#### (b) Which 5 counties have the lowest (most negative) log-likelihood values? Please show a ranked list of the top counties' names and log-likelihood values. (10%)

```python
from pgmpy.models import BayesianNetwork
from pgmpy.estimators import MaximumLikelihoodEstimator

model = BayesianNetwork(best_model.edges())
model.fit(Data3, estimator=MaximumLikelihoodEstimator)
for cpd in model.get_cpds():
    print("CPD of {variable}:".format(variable=cpd.variable))
    print(cpd)
```

```python
# Log-likelihood of data record x given model
def LL(x,model,verbose=False):
    loglike = 0
    for cpd in model.get_cpds():
        temp_cpd = cpd.copy()
        thevariable = temp_cpd.variable
        theparents = model.predecessors(thevariable)
        theparents_list = []
        for parent in theparents:
            temp_cpd.reduce([(parent, x[parent])])
            theparents_list.append(parent)
        theprob = temp_cpd.get_values()[x[thevariable],0]
        if verbose:
            print(thevariable,theparents_list,theprob)
        loglike += np.log(theprob)
    return loglike
```

```python
Data3=Data3.dropna()
```

```python
Data3['ll']=Data3.iloc[:, 1:].apply(lambda x: LL(x,model),axis=1)
Data3_sorted = Data3.sort_values(by='ll')
top_5_lowest_ll = Data3_sorted[['County', 'll']].head()
top_5_lowest_ll
```

Cluster based anomaly detection. Use all numeric features for this part, and do not discretize.

#### (a) Clustering with k-means. Please use k=3 clusters. Compute each record's distance to the nearest cluster center and report the five counties which have the longest distances. (10%)

```python
Data2=pd.read_csv("CountyHealthIndicators.csv")
```

```python
Data2 = Data2.replace('?', np.nan)
Data2 = Data2.dropna()
```

```python
Data2_Clean=Data2.iloc[:, 1:].replace('?', np.nan).apply(pd.to_numeric, errors='coerce')
Data2_Clean = Data2_Clean.dropna()
```

```python
from sklearn.cluster import KMeans
import pandas as pd

# Initialize and fit KMeans
KM = KMeans(n_clusters=3, random_state=999)
KM.fit(Data2_Clean)

# Compute distances to cluster centers and select the nearest one
distances = KM.transform(Data2_Clean)
Data2['top_distances'] = pd.DataFrame({
    'score': distances.min(axis=1)
})

# Sort and display the top 5 records with the highest distance to nearest cluster center
```

```python
top_five_distances =  Data2[['County', 'top_distances']].sort_values(by='top_distances',ascending=False).head(5)

top_five_distances
```

```python
from sklearn.mixture import GaussianMixture

# Initialize and fit KMeans
GM=GaussianMixture(n_components=3,random_state=999)
GM.fit(Data2_Clean)

# Compute distances to cluster centers and select the nearest one
Data2['GMM_top_distances'] = GM.score_samples(Data2_Clean)
# Sort and display the top 5 records with the highest distance to nearest cluster center
top_five_distances = Data2[['County', 'GMM_top_distances']].sort_values(by='GMM_top_distances').head(5)

top_five_distances
```
